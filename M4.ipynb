```python
import cv2
from cvzone.HandTrackingModule import HandDetector
from cvzone.ClassificationModule import Classifier
import numpy as np
import math
import pyautogui
import time
 
cap = cv2.VideoCapture(0)
detector = HandDetector(maxHands=1)
classifier = Classifier("model/keras_model.h5")
count=0
 
offset = 20
imgSize = 224
 
labels = ["Callme", "Ok", 'Peace','Rock','Thumpsdown','Thumpsup']
count=0
select1=0
select2=0
select3=0
select4=0
select5=0



while True:
    success, img = cap.read()

    imgOutput = img.copy()

    try:
        hands, img = detector.findHands(img)
        if hands:
            hand = hands[0]
            x, y, w, h = hand['bbox']

            imgWhite = np.ones((imgSize, imgSize, 3), np.uint8) * 255
            imgCrop = img[y - offset:y + h + offset, x - offset:x + w + offset]

            imgCropShape = imgCrop.shape

            aspectRatio = h / w

            if aspectRatio > 1:
                
                k = imgSize / h
                wCal = math.ceil(k * w)
                imgResize = cv2.resize(imgCrop, (wCal, imgSize))
                imgResizeShape = imgResize.shape
                wGap = math.ceil((imgSize - wCal) / 2)
                imgWhite[:, wGap:wCal + wGap] = imgResize
                prediction, index = classifier.getPrediction(imgWhite, draw=False)
                #print(prediction, index)

            else:
                k
                k = imgSize / w
                hCal = math.ceil(k * h)
                imgResize = cv2.resize(imgCrop, (imgSize, hCal))
                imgResizeShape = imgResize.shape
                hGap = math.ceil((imgSize - hCal) / 2)

                imgWhite[hGap:hCal + hGap, :] = imgResize
                
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
            prediction, index = classifier.getPrediction(imgWhite, draw=False)
            
            cv2.rectangle(imgOutput, (x - offset, y - offset-50),
                          (x - offset+90, y - offset-50+50), (255, 0, 255), cv2.FILLED)
            if labels[index]!='Callme':
                cv2.putText(imgOutput, labels[index], (x, y -26), cv2.FONT_HERSHEY_COMPLEX, 1.7, (255, 255, 255), 2)
            cv2.rectangle(imgOutput, (x-offset, y-offset),
                          (x + w+offset, y + h+offset), (255, 0, 255), 4)
            print(labels[index])
            
            
            ####################################################
            if labels[index]=='Ok':
                select1+=1
                
            if labels[index]=='Thumpsup':
                select2+=1
                
            if labels[index]=='Thumpsdown':
                select3+=1
            
            
            if labels[index]=='Peace':
                select4+=1  
                
            if labels[index]=='Rock':
                select5+=1
                
                
                
            
            if labels[index]=='Ok' and count==0 and select1>5:     # Pause

                pyautogui.press('space')
                count=1

                time.sleep(2)
                select1=0

            elif labels[index]=='Ok' and count==1 and select1>5:   # Play

                pyautogui.press('space')
                count=0
                time.sleep(2)
                select1=0

            
            if labels[index]=='Thumpsup' and select2>5:
                

                pyautogui.press('n')
                time.sleep(2)
                select2=0
            
            if labels[index]=='Thumpsdown' and select3>5:               

                
                pyautogui.press('p')
                time.sleep(2)
                select3=0
                
            if labels[index]=='Peace' and select4>3 :
                
                #time.sleep(2)
                pyautogui.press('right')
                select4=0
                
            if labels[index]=='Rock' and select5>3 :
                
                pyautogui.press('left')
                select5=0

                

            
    except:
        print('not detected') 

    cv2.imshow("Image", imgOutput)
    k = cv2.waitKey(1)
    if k == 27:
        break

cv2.destroyAllWindows()
```
